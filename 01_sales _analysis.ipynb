```python
path = r'dataset.xlsx' # path to the dataset
xls = pd.ExcelFile(path) 

sheet_names = [s for s in xls.sheet_names if s != "Customers"]

Customers = pd.read_excel(path, sheet_name="Customers")

Customers = pd.merge(
    Customers,
    pd.read_excel(path, sheet_name="Regions"),
    left_on="Region_ID",
    right_on="Region_ID",
    how="left"
)
master = pd.read_excel(path, sheet_name="Orders")


FK_IDs = ['Product_ID', 'Sales_Rep_ID', 'Region_ID']

for i in range(len(FK_IDs)):
    detail = pd.read_excel(path, sheet_name=sheet_names[i+1])  # +1 لأن Orders هو master
    master = pd.merge(master, detail, left_on=FK_IDs[i], right_on=FK_IDs[i], how="left")

master = pd.merge(
    master,
    Customers[['Customer_ID', 'Customer_Name' ,'Region_ID',]],
    left_on="Region_ID",
    right_on="Region_ID",
    how="left"
)
```

```python

df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Period'] = df['Order_Date'].dt.to_period('M')


monthly = (
    df
    .groupby('Period', as_index=True)
    .agg(Total_Sales=('Total_Sales', 'sum'))
    .sort_index()
    .reset_index()
)
monthly['PeriodStart'] = monthly['Period'].dt.to_timestamp()
monthly['MoM_pct'] = monthly['Total_Sales'].pct_change(periods=1) * 100
monthly['YoY_pct'] = monthly['Total_Sales'].pct_change(periods=12) * 100

fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Bar(
        x=monthly['PeriodStart'],
        y=monthly['Total_Sales'],
        name='Total Sales',
        marker_color='rgba(58, 71, 80, 0.8)',
        hovertemplate='%{x|%Y-%m}<br>Total Sales: %{y:,.2f}$<extra></extra>'
    ),
    secondary_y=False
)

# خط MoM
fig.add_trace(
    go.Scatter(
        x=monthly['PeriodStart'],
        y=monthly['MoM_pct'],
        mode='lines+markers',
        name='MoM %',
        line=dict(color='orange', width=3),
        marker=dict(size=8),
        hovertemplate='%{x|%Y-%m}<br>MoM Growth: %{y:.2f}%<extra></extra>'
    ),
    secondary_y=True
)

fig.add_trace(
    go.Scatter(
        x=monthly['PeriodStart'],
        y=monthly['YoY_pct'],
        mode='lines+markers',
        name='YoY %',
        line=dict(color='red', width=3, dash='dash'),
        marker=dict(size=8),
        hovertemplate='%{x|%Y-%m}<br>YoY Growth: %{y:.2f}%<extra></extra>'
    ),
    secondary_y=True
)

fig.add_hline(y=0, line=dict(color='black', width=1, dash='dot'), secondary_y=True)

fig.update_layout(
 title=dict(
    text='Monthly Sales with MoM and YoY Growth (%)',
        x=0.5,                
        xanchor='center',
        font=dict(size=30)
    ),
    xaxis=dict(title='Month'),
    yaxis=dict(title='Total Sales', showgrid=True),
    yaxis2=dict(title='Growth (%)', showgrid=False),
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
    hovermode='x unified',
    template='plotly_dark',
    bargap=0.2,
    margin=dict(l=60, r=60, t=80, b=60)
    
)

fig.update_yaxes(tickformat=".1f", secondary_y=True)
fig.show()

```

# Sales Insights (Oct 2022 – Oct 2024)

---

## 1. Initial Growth Phase (Oct–Dec 2022)
- Sales jumped sharply from **646K (Oct 2022)** to **4.13M (Dec 2022)**.
- **Nov 2022** saw the strongest monthly growth (**+387% MoM**) — likely due to a **seasonal event (promotions, holidays, or launch activity)**.

---

## 2. Fluctuations in 2023
- After peaking in **Dec 2022**, sales **dropped 25% in Jan 2023** and **16% in Feb 2023**.
- 2023 showed **alternating growth and decline cycles**, suggesting sensitivity to **campaigns or external factors**.
- **Sep 2023** hit **3.57M (+8.7% MoM)**, a strong month.
- **Oct 2023** dropped -23% MoM but showed **+325% YoY vs Oct 2022**.
- Overall: **volatile but upward trend compared to 2022**.

---

## 3. Performance in 2024
- **Jan 2024:** Weak start (**-25.5% MoM, -10.6% YoY**).
- **Feb 2024:** Strong rebound (**+18% MoM, +25.8% YoY**).
- **Mar 2024:** Sharp decline (**-26% MoM, -25% YoY**) → one of the weakest months.
- **Apr–Jun 2024:** Strong recovery  
  - Apr: **+35% MoM, +13.5% YoY**  
  - Jun: **3.77M (+10% MoM, +13% YoY)** → one of the best months in 2024.
- **Jul–Sep 2024:** Downturn  
  - Jul: **-17.6% MoM, +1.6% YoY**  
  - Sep: **-20.8% MoM, -27.7% YoY** → major concern.
- **Oct 2024:** Slight recovery (**+6% MoM**) but **flat YoY (-0.4%)**.

---

## 4. Key Patterns
- **Seasonality:** Strong **Nov–Dec peaks** (holiday/end-of-year).
- **Q1 Weakness:** Jan–Mar generally weaker (seen in 2023 & 2024).
- **Mid-Year Recovery:** Apr–Jun rebounds strongly.
- **Volatility:** Sales swings heavily → driven by **campaigns/events**.
- **2024 Warning Signs:** H1 was strong, but **Q3 saw steep YoY declines**.

---

## 5. Actionable Insights
- **Maximize Q4 campaigns** (Nov–Dec) to capture peak growth.
- **Reduce Q1 slumps** with **new campaigns or retention strategies**.
- **Investigate Q3 2024 drop** → possible causes:  
  - Market saturation  
  - Competitor activity  
  - Reduced promotions
- **Stabilize growth** → use **loyalty programs, consistent offers, and new channels** to reduce reliance on seasonal spikes.

```python

order_summary = df.groupby('Order_ID').agg({
    'Total_Sales':'sum',
    'Quantity_Sold':'sum'
}).reset_index()

order_summary['Average_Order_Size'] = order_summary['Total_Sales'] / order_summary['Quantity_Sold']

mean_order_size = order_summary['Average_Order_Size'].mean()
median_order_size = order_summary['Average_Order_Size'].median()

print(f"Average Order Size (Mean): ${mean_order_size:.2f}")
print(f"Average Order Size (Median): ${median_order_size:.2f}")
```

```python


region_summary = df.groupby('Region_Name').agg({
    'Total_Sales':'sum',
    'Quantity_Sold':'sum'
}).reset_index()
region_summary['Average_Order_Size'] = region_summary['Total_Sales'] / region_summary['Quantity_Sold']

fig = px.bar(
    region_summary.round(2),
    x='Region_Name',
    y='Average_Order_Size',
    # text='Average_Order_Size',           
    labels={'Average_Order_Size':'Average Order Size ($)', 'Region_Name':'Region'},
    color_discrete_sequence=['#34ebd8']
)

fig.update_layout(
    title='Average Order Size per Region',
    title_x=0.5,
    template='plotly_dark'
)

region_summary
```

```python

fig.show()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
monthly_summary = df.groupby(df['Order_Date'].dt.to_period('M')).agg({
    'Total_Sales':'sum',
    'Quantity_Sold':'sum'
}).reset_index().round(2)
monthly_summary['Average_Order_Size'] = monthly_summary['Total_Sales'] / monthly_summary['Quantity_Sold']
monthly_summary['Period'] = monthly_summary['Order_Date'].dt.strftime('%b %Y')

fig = px.line(
    monthly_summary.round(2),
    x='Period',
    y='Average_Order_Size',
    markers=True,
    text='Average_Order_Size',
    labels={'Average_Order_Size':'Average Order Size ($)', 'Period':'Month'}
)

fig.update_layout(
    title='Monthly Average Order Size',
    title_x=0.5,
    template='plotly_dark',
    xaxis_tickangle=-45
)

monthly_summary
```

```python

df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce').dt.hour
df['Year_Month'] = df['Year'].astype(str) + '-' + df['Month'].astype(str).str.zfill(2)

def get_time_of_day(hour):
    if pd.isna(hour):
        return "Unknown"
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

df['Time_of_Day'] = df['Hour'].apply(get_time_of_day)

hour_sales = df.groupby('Hour', as_index=False)['Total_Sales'].sum()
time_sales = df.groupby('Time_of_Day', as_index=False)['Total_Sales'].sum()
time_order = ['Morning','Afternoon','Evening','Night','Unknown']
time_sales['Time_of_Day'] = pd.Categorical(time_sales['Time_of_Day'], categories=time_order, ordered=True)
time_sales = time_sales.sort_values('Time_of_Day')
monthly_sales = df.groupby(['Year','Month'], as_index=False)['Total_Sales'].sum()
monthly_sales['Year_Month'] = monthly_sales['Year'].astype(str) + '-' + monthly_sales['Month'].astype(str).str.zfill(2)
yearly_sales = df.groupby('Year', as_index=False)['Total_Sales'].sum()

fig = make_subplots(
    rows=2, cols=2,
    specs=[[{"type":"bar"}, {"type":"pie"}],
           [{"type":"scatter"}, {"type":"scatter"}]],
    subplot_titles=("Hourly Sales Distribution", "Sales by Time of Day",
                    "Monthly Sales Trend", "Yearly Sales Trend")
)

colors_hour = ['#34c0eb' if h < 12 else '#0a9396' if h < 17 else '#f4d35e' if h < 21 else '#ee6c4d' for h in hour_sales['Hour']]
fig.add_trace(
    go.Bar(
        x=hour_sales['Hour'],
        y=hour_sales['Total_Sales'],
        textposition='outside',
        marker_color=colors_hour,
        name='Hourly Sales',
        hovertemplate='Hour: %{x}<br>Sales: $%{y:,.0f}<extra></extra>'
    ),
    row=1, col=1
)
fig.update_xaxes(title_text="Hour of Day", row=1, col=1)
fig.update_yaxes(title_text="Total Sales ($)", row=1, col=1)

fig.add_trace(
    go.Pie(
        labels=time_sales['Time_of_Day'],
        values=time_sales['Total_Sales'],
        hole=0.4,
        textinfo='label+percent',
        marker=dict(colors=['#34c0eb','#0a9396','#0077b6','#f4d35e','#ee6c4d']),
        name='Time of Day Sales',
        hovertemplate='%{label}: $%{value:,.0f} (%{percent})<extra></extra>'
    ),
    row=1, col=2
)

for year in monthly_sales['Year'].unique():
    df_year = monthly_sales[monthly_sales['Year']==year]
    fig.add_trace(
        go.Scatter(
            x=df_year['Year_Month'],
            y=df_year['Total_Sales'],
            mode='lines+markers+text',
            textposition='top center',
            name=str(year),
            line=dict(width=3),
            marker=dict(size=8),
            hovertemplate='Month: %{x}<br>Sales: $%{y:,.0f}<extra></extra>'
        ),
        row=2, col=1
    )
fig.update_xaxes(title_text="Month", row=2, col=1)
fig.update_yaxes(title_text="Total Sales ($)", row=2, col=1)

fig.add_trace(
    go.Scatter(
        x=yearly_sales['Year'],
        y=yearly_sales['Total_Sales'],
        mode='lines+markers+text',
        textposition='top center',
        line=dict(color='#f4d35e', width=4),
        marker=dict(size=12, color='#ee6c4d'),
        name='Yearly Sales',
        hovertemplate='Year: %{x}<br>Sales: $%{y:,.0f}<extra></extra>'
    ),
    row=2, col=2
)
fig.update_xaxes(title_text="Year", row=2, col=2)
fig.update_yaxes(title_text="Total Sales ($)", row=2, col=2)

fig.update_layout(
    title_text="📊 Sales Analysis: Hourly, Time of Day, Monthly & Yearly Trends",
    title_x=0.5,
    template='plotly_dark',
    font=dict(size=14),
    showlegend=True
)

fig.show()
```
